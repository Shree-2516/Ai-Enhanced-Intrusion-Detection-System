Cell 1: Imports and Loading


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load your final processed data
df = pd.read_csv('../dataset/processed/final_encoded_data.csv')
print(f"Dataset shape: {df.shape}")
df.head()
```

Cell 2: Attack Distribution (Class Imbalance Check)

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='Label', data=df)
plt.title('Distribution of Traffic Labels')
plt.xticks(rotation=45)
plt.show()
```

Cell 3: Feature Correlation Heatmap


```python
plt.figure(figsize=(15, 10))
# Selecting a subset of numeric columns to avoid a cluttered heatmap
corr = df.select_dtypes(include=['float64', 'int64']).corr()
sns.heatmap(corr, cmap='coolwarm', annot=False)
plt.title('Feature Correlation Heatmap')
plt.show()
```

Cell 4: Visualizing a Specific Feature vs. Label

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Label', y='Destination Port', data=df)
plt.title('Destination Port Distribution by Label')
plt.xticks(rotation=45)
plt.show()
```
